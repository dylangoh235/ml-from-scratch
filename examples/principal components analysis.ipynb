```python
import numpy as np
import sklearn.datasets as dataset
from sklearn.model_selection import train_test_split
from machine_learning.unsupervised_learning.principal_components_analysis import PCA
import matplotlib.pyplot as plt
```

```python
digits = dataset.load_digits()
n_samples = len(digits.images)
data = digits.images.reshape((n_samples, -1))

X_train, X_test, y_train, y_test = train_test_split(data, digits.target, test_size=0.3, shuffle=False)
```

```python
pca = PCA(r=3)
y_pred = pca.transform(X_train)
```

```python
colors = plt.cm.jet(plt.Normalize()(y_train))

plt.scatter(y_pred[:,0], y_pred[:,1],c=colors)
plt.xlabel("PCA1")
plt.ylabel("PCA2")
plt.show()

plt.scatter(y_pred[:,1], y_pred[:,2],c=colors)
plt.xlabel("PCA2")
plt.ylabel("PCA3")
plt.show()
```

```python
# make explained_variance_ratio_
plt.plot(np.cumsum(pca.explained_variance_ratio_), 'o-')
plt.xlabel("Number of Components")
plt.ylabel("Cumulative variance retained")
plt.show()
```
