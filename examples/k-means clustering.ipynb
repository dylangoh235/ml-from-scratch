```python
import numpy as np
import sklearn.datasets as dataset
from sklearn.model_selection import train_test_split
from machine_learning.unsupervised_learning.k_means_clustering import KMeans
from machine_learning.utils import *
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
X, y = dataset.make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=y)
plt.show()
```

```python
clf = KMeans(4, max_iter=1000)
clf.transform(X_train)
```

```python
sns.scatterplot(x=X_train[:, 0], y=X_train[:, 1], hue=y_train)
for i in range(clf.k):
    plt.scatter(*clf.centroids[i, :])
plt.show()
```

```python
sns.scatterplot(x=X_test[:, 0], y=X_test[:, 1], hue=y_test)
for i in range(clf.k):
    plt.scatter(*clf.centroids[i, :])
plt.show()
```
