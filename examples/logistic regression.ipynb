```python
import numpy as np
import sklearn.datasets as datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from machine_learning.supervised_learning.logistic_regression import Logistic
from machine_learning.utils.regularization import *
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
X, y = datasets.make_blobs(n_samples=100, centers=2, random_state=6)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, shuffle=False)

y_train = y_train.reshape(-1,1)
```

```python
logistic = Logistic()
loss = logistic.train(X_train, y_train, epochs=7)
y_hat = logistic.predict(X_test)

plt.plot(loss)
plt.show()
```

```python
sns.scatterplot(
    x=X[:, 0], y=X[:, 1], hue=y, marker="o", s=25, edgecolor="k"
)
plt.show()

sns.scatterplot(
    x=X_test[:, 0], y=X_test[:, 1], hue=y_hat.reshape(-1), marker="o", s=25, edgecolor="k"
)

plt.show()
```
